```python
# Rainfall data for the year (in mm)
rainfall = [45, 52, 120, 156, 180, 165, 140, 135, 98, 65, 30, 28]

# List of months
months = ['January', 'February', 'March', 'April', 'May', 'June',
          'July', 'August', 'September', 'October', 'November', 'December']

# 1. Calculate total annual rainfall using a for loop
total_rainfall = 0
for r in rainfall:
    total_rainfall += r

# 2. Calculate average monthly rainfall
average_rainfall = total_rainfall / 12

# 3. Find maximum and minimum rainfall and their months
max_rainfall = max(rainfall)
min_rainfall = min(rainfall)
max_index = rainfall.index(max_rainfall)
min_index = rainfall.index(min_rainfall)
max_month = months[max_index]
min_month = months[min_index]

# 4. Find months with above-average rainfall
above_avg_months = []
for i in range(len(rainfall)):
    if rainfall[i] > average_rainfall:
        above_avg_months.append(months[i])

# Number of such months
num_above_avg = len(above_avg_months)

# Output the analysis
print("Rainfall Analysis for 2025:")
print(f"Total Annual Rainfall: {total_rainfall} mm")
print(f"Average Monthly Rainfall: {average_rainfall:.2f} mm")
print(f"Maximum Rainfall: {max_rainfall} mm ({max_month})")
print(f"Minimum Rainfall: {min_rainfall} mm ({min_month})")
print(f"Months with above-average rainfall: {', '.join(above_avg_months)}")
print(f"Number of such months: {num_above_avg}")
```

```python
# Maximum possible yield
max_yield = 50

# Initialize variables
week = 1
percentage = 5

# While loop with conditional exit when yield reaches 80% of maximum
while True:
    # Calculate yield estimate
    yield_est = max_yield * percentage / 100
    # Print the weekly report
    print(f"Week {week}: Yield estimate = {yield_est:.1f} quintals/ha ({percentage:.1f}% of max)")
    # Check if reached 80%
    if percentage >= 80:
        break
    # Update percentage based on week
    if week == 1:
        percentage = 12
    elif week == 2:
        percentage = 25
    else:
        percentage += 5
    # Increment week
    week += 1

# Final message
print("\nCROP READY FOR HARVEST!")
print(f"Total weeks to harvest maturity: {week} weeks")
```

```python
# Irrigation schedule for 20 days, every 3rd day starting from day 1
irrigation_days = []
total_water = 0
irrigation_amount = 50  # mm per irrigation

# Use a for loop to check each day from 1 to 20
for day in range(1, 21):
    if (day - 1) % 3 == 0:  # Irrigation on day 1, 4, 7, 10, 13, 16, 19
        irrigation_days.append(day)
        total_water += irrigation_amount

# Output the schedule
print("20-Day Irrigation Schedule (every 3rd day):")
for day in irrigation_days:
    print(f"Day {day}: Irrigate {irrigation_amount}mm")

# Calculate and print totals
total_days = len(irrigation_days)
print(f"Total irrigation days: {total_days}")
print(f"Total water applied: {total_water}mm")
```

```python
# Pest monitoring weekly records
# Example pest counts (you can replace with actual data)
pest_counts = [5, 8, 12, 18, 22, 28]  # Pest counts for weeks 1 to 6

# Threshold for pest count
threshold = 15

# Initialize variables
week = 1
increases = []

# While loop to analyze pest growth trend
while week <= len(pest_counts):
    count = pest_counts[week - 1]

    if week > 1:
        # Calculate week-to-week increase
        increase = count - pest_counts[week - 2]
        increases.append(increase)
        print(f"Week {week}: Pest count = {count}, Increase = {increase}")
    else:
        print(f"Week {week}: Pest count = {count}")

    # Check if threshold exceeded
    if count > threshold:
        print(f"Alert: Pest count exceeded threshold of {threshold} at week {week}")
        break

    week += 1

# Running calculation: Average increase and prediction
if increases:
    avg_increase = sum(increases) / len(increases)
    print(f"\nAverage week-to-week increase: {avg_increase:.2f}")

    # Predict next week's pest count
    predicted_count = count + avg_increase
    print(f"Predicted pest count next week: {predicted_count:.2f}")

    if predicted_count > threshold:
        print("Control measures needed next week.")
else:
    print("No increases calculated yet.")
```

```python
# Harvest yield recording for 5 farmers
yields = [45, 52, 48, 55, 50]  # Yields in quintals

# Initialize variables for accumulation
total_yield = 0
highest_yield = yields[0]
lowest_yield = yields[0]

# For loop to accumulate total and find highest/lowest
for yield_val in yields:
    total_yield += yield_val
    if yield_val > highest_yield:
        highest_yield = yield_val
    if yield_val < lowest_yield:
        lowest_yield = yield_val

# Calculate average yield
average_yield = total_yield / len(yields)

# Output the results
print("Harvest Yield Analysis:")
print(f"Total yield: {total_yield} quintals")
print(f"Average yield: {average_yield:.2f} quintals")
print(f"Highest yield: {highest_yield} quintals")
print(f"Lowest yield: {lowest_yield} quintals")
```

```python
# Daily maximum temperatures for 15 days (sample data in °C)
temperatures = [25, 28, 32, 30, 35, 38, 40, 42, 39, 37, 35, 33, 31, 29, 27]

# Initialize variables
days_above_30 = []
max_temp = temperatures[0]
min_temp = temperatures[0]
increases = 0
decreases = 0

# For loop with condition checking
for i in range(len(temperatures)):
    temp = temperatures[i]

    # Check for days > 30°C
    if temp > 30:
        days_above_30.append(i + 1)  # Day number (1-based)

    # Update max and min for range
    if temp > max_temp:
        max_temp = temp
    if temp < min_temp:
        min_temp = temp

    # Check trend: compare with previous day (if not first day)
    if i > 0:
        if temp > temperatures[i-1]:
            increases += 1
        elif temp < temperatures[i-1]:
            decreases += 1

# Calculate temperature range
temp_range = max_temp - min_temp

# Determine overall trend
if increases > decreases:
    trend = "Increasing"
elif decreases > increases:
    trend = "Decreasing"
else:
    trend = "Stable"

# Output the analysis
print("Temperature Trend Analysis:")
print(f"Days with temperature > 30°C: {', '.join(map(str, days_above_30))}")
print(f"Temperature range: {temp_range}°C (Max: {max_temp}°C, Min: {min_temp}°C)")
print(f"Overall trend: {trend} (Increases: {increases}, Decreases: {decreases})")
```

```python
# Soil moisture monitoring simulation
# Initial moisture: 75%
# Evaporation rate: 2.5% per hour
# Irrigation threshold: 40%
# Irrigation restores to 80%

# Initialize variables
moisture = 75.0
evaporation_rate = 2.5
threshold = 40.0
irrigation_level = 80.0
hour = 0
max_hours = 24  # Simulate for 24 hours
irrigation_count = 0

# While loop with threshold checking
while hour < max_hours:
    hour += 1
    moisture -= evaporation_rate

    # Print current moisture
    print(f"Hour {hour}: Moisture = {moisture:.1f}%")

    # Check if irrigation is needed
    if moisture <= threshold:
        print(f"Irrigation triggered at Hour {hour}! Moisture restored to {irrigation_level}%")
        moisture = irrigation_level
        irrigation_count += 1

# Summary
print(f"\nTotal irrigations triggered in {max_hours} hours: {irrigation_count}")
```

```python
# Crop growth stages with day ranges
stages = [
    ("Germination", 0, 7),
    ("Seedling", 8, 21),
    ("Vegetative", 22, 45),
    ("Flowering", 46, 65),
    ("Grain filling", 66, 85),
    ("Maturity", 86, float('inf'))  # Maturity has no end
]

# Function to determine stage and days remaining
def get_growth_status(current_day):
    current_stage = None
    days_remaining = 0

    # For loop with status update: iterate through stages to find current one
    for stage, start, end in stages:
        if start <= current_day <= end:
            current_stage = stage
            if end == float('inf'):
                days_remaining = "Indefinite (Maturity reached)"
            else:
                days_remaining = end - current_day
            break  # Exit loop once stage is found

    return current_stage, days_remaining

# Example usage: Assume current day is 50 (you can change this or take input)
current_day = 50  # Example; replace with int(input("Enter current day: ")) for user input

# Get status
stage, remaining = get_growth_status(current_day)

# Output
print(f"Current Day: {current_day}")
print(f"Growth Stage: {stage}")
print(f"Days Remaining in Stage: {remaining}")
```

```python
# Simple Cost-Benefit Analysis for 5 years
# Conventional: ₹50k invest, ₹80k/year yield
# Organic: ₹70k invest, ₹95k/year yield, ₹5k fee year 3

conv_profit = 0
org_profit = 0

for year in range(1, 6):
    conv_cost = 50000 if year == 1 else 0
    org_cost = 70000 if year == 1 else 0
    if year == 3:
        org_cost += 5000
    conv_profit += 80000 - conv_cost
    org_profit += 95000 - org_cost

print(f"Conventional 5-year profit: ₹{conv_profit}")
print(f"Organic 5-year profit: ₹{org_profit}")
if org_profit > conv_profit:
    print("Organic is better.")
else:
    print("Conventional is better.")
```
