```python
age=int(input("enter age"))
if age <=6:
  print("no ticket")
elif age>6 and age <=12:
  print("half ticket")
else:
  print("full ticket")
```

```python
marks = int(input("enter marks"))
if marks >=90:
  print("A grade")
elif marks >=80 and marks <90:
  print("B grade")
elif marks >=70 and marks <80:
  print("C grade")
```

```python
cgpa=float(input("enter cgpa"))
if cgpa > 8:
  print("eligible for placement")
else:
  has_backlogs_input = input("Do you have any backlogs? (yes/no): ").lower()
  no_backlogs = (has_backlogs_input == 'no')
  current_semester = int(input("Enter your current semester: "))
  if no_backlogs and (current_semester == 8):
    print("eligible for placement")
  else:
    print("not eligible for placement")
```
