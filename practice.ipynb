```python
crop = input("Enter the crop (rice, wheat, sugarcane): ").lower()
moisture = float(input("Enter soil moisture percentage: "))
temp = input("Enter temperature condition (hot, moderate, cool): ").lower()

if crop == "rice":
    if (temp == "hot" and moisture < 60) or (temp in ["moderate", "cool"] and moisture < 50):
        print("Rice needs water.")
    else:
        print("Rice does not need water.")

elif crop == "wheat":
    if (temp == "hot" and moisture < 40) or (temp in ["moderate", "cool"] and moisture < 50):
        print("Wheat needs water.")
    else:
        print("Wheat does not need water.")

elif crop == "sugarcane":
    if (temp == "hot" and moisture < 70) or (temp in ["moderate", "cool"] and moisture < 60):
        print("Sugarcane needs water.")
    else:
        print("Sugarcane does not need water.")

else:
    print("Invalid crop type entered.")
```
