```python
marks=[56,78,90,45,88]
#Max
max_marks=max(marks)

#average
total=0
for m in marks:
  total=total+m
average=total/len(marks)

#count>=60
count=0
for m in marks:
  if m>=60:
    count=count+m
print("max:",max_marks)
print("average:",average)
print("count>=60:",count)
```
