```python
# Simple Python program for precision irrigation system

crop = input("Enter crop type (rice, wheat, sugarcane): ").lower()
moisture = float(input("Enter soil moisture percentage: "))
temp = input("Enter temperature condition (hot, moderate, cool): ").lower()

if crop == "rice":
    if temp == "hot" and moisture < 60:
        print("Irrigation needed for rice.")
    elif (temp == "moderate" or temp == "cool") and moisture < 50:
        print("Irrigation needed for rice.")
    else:
        print("No irrigation needed for rice.")
elif crop == "wheat":
    if temp == "hot" and moisture < 40:
        print("Irrigation needed for wheat.")
    elif (temp == "moderate" or temp == "cool") and moisture < 30:
        print("Irrigation needed for wheat.")
    else:
        print("No irrigation needed for wheat.")
elif crop == "sugarcane":
    if temp == "hot" and moisture < 70:
        print("Irrigation needed for sugarcane.")
    elif (temp == "moderate" or temp == "cool") and moisture < 60:
        print("Irrigation needed for sugarcane.")
    else:
        print("No irrigation needed for sugarcane.")
else:
    print("Invalid crop type.")
```

```python
# Simple Python program for crop monitoring system pest severity classification

pests = int(input("Enter number of pests per plant: "))
damage = float(input("Enter crop damage percentage: "))
temp = input("Enter temperature (hot, moderate, cool): ").lower()
humidity = float(input("Enter humidity level (%): "))

# Classify severity
if pests < 5 and damage < 10:
    severity = "Low"
elif (5 <= pests <= 15) or (10 <= damage <= 25):
    severity = "Medium"
else:
    severity = "High"

# Recommend action based on severity and conditions
if severity == "Low":
    action = "No action needed. Monitor regularly."
elif severity == "Medium":
    action = "Apply organic pesticide."
else:  # High
    action = "Apply chemical pesticide."
    if temp == "hot" and humidity < 50:
        action += " Increase irrigation to reduce pest reproduction."

print(f"Pest severity: {severity}")
print(f"Recommended action: {action}")
```

```python
# Simple Python program to check crop harvest readiness

crop = input("Enter crop type (rice, wheat, cotton): ").lower()
days = int(input("Enter days since sowing: "))

if crop == "rice":
    maturity_min, maturity_max = 120, 150
elif crop == "wheat":
    maturity_min, maturity_max = 140, 180
elif crop == "cotton":
    maturity_min, maturity_max = 160, 200
else:
    print("Invalid crop type.")
    exit()

if maturity_min <= days <= maturity_max:
    status = "READY FOR HARVEST"
    recommendation = "Begin harvesting within next 2 weeks for optimal yield."
    yield_loss = "2% per day"
else:
    status = "NOT READY FOR HARVEST"
    recommendation = "Wait for the crop to mature."
    yield_loss = "N/A"

print(f"Crop: {crop.capitalize()}")
print(f"Days since sowing: {days}")
print(f"Status: {status}")
print(f"Maturity range: {maturity_min}-{maturity_max} days")
print(f"Recommendation: {recommendation}")
print(f"Expected yield loss if delayed: {yield_loss}")
```

```python
# Simple Python program to classify soil pH and recommend amendments

ph = float(input("Enter soil pH value: "))

if ph < 5.5:
    classification = "Highly acidic"
    recommendation = "Needs lime."
elif 5.5 <= ph < 6.5:
    classification = "Acidic"
    recommendation = "May need amendment."
elif 6.5 <= ph < 7.5:
    classification = "Neutral"
    recommendation = "Ideal."
elif 7.5 <= ph < 8.5:
    classification = "Alkaline"
    recommendation = "Needs sulfur."
else:  # ph >= 8.5
    classification = "Highly alkaline"
    recommendation = "Needs major amendment."

print(f"Soil classification: {classification}")
print(f"Recommendation: {recommendation}")
```

```python
# Simple Python program for crop yield prediction

base_yield = 40  # quintals per hectare

rainfall = float(input("Enter rainfall (mm): "))
temperature = float(input("Enter temperature (°C): "))
soil_quality = float(input("Enter soil quality (0-100 points): "))

# Rainfall adjustment
if rainfall > 600:
    rainfall_adj = (rainfall - 600) / 10 * 0.02
elif rainfall < 400:
    rainfall_adj = (400 - rainfall) / 10 * -0.02
else:
    rainfall_adj = 0

# Temperature adjustment
if temperature < 25:
    temp_adj = (25 - temperature) * -0.01
elif temperature > 30:
    temp_adj = (temperature - 30) * -0.01
else:
    temp_adj = 0

# Soil quality adjustment (proportional multiplier)
soil_multiplier = soil_quality / 100

# Calculate final yield
final_yield = base_yield * (1 + rainfall_adj + temp_adj) * soil_multiplier

print(f"Predicted yield: {final_yield:.2f} quintals per hectare")
```

```python
# Simple Python program for disease risk assessment

temperature = float(input("Enter temperature (°C): "))
humidity = float(input("Enter humidity (%): "))
crop_age = int(input("Enter crop age (days): "))
history = input("Previous season disease history (yes/no): ").lower()
susceptibility = int(input("Enter variety susceptibility rating (1=low, 2=medium, 3=high): "))

temp_score = 3 if 20 <= temperature <= 30 else 1
humidity_score = 3 if humidity > 80 else 2 if humidity >= 50 else 1
age_score = 3 if crop_age < 30 else 2 if crop_age <= 90 else 1
history_score = 3 if history == "yes" else 1
suscep_score = susceptibility

total_score = temp_score + humidity_score + age_score + history_score + suscep_score

risk = "Low" if total_score < 7 else "Medium" if total_score <= 10 else "High"

print(f"Total risk score: {total_score}")
print(f"Overall disease risk: {risk}")
```
