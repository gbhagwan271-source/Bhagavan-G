```python
import pandas as pd

data = {
"Rainfall_mm": [180, 190, 200, 210, 220],
"Yield_kg": [3000, 3100, 3200, 3400, 3500],
"fertilizer_kg": [40,45,50,55,60]
}

df = pd.DataFrame(data)
```

```python
from sklearn.linear_model import LinearRegression

X = df[["Rainfall_mm"]] # Input: rainfall
y = df["Yield_kg"] # Output: yield

model = LinearRegression()
model.fit(X, y)
```

```python
# Predict yield for 205 mm rainfall
model.predict([[205]])
```

```python
model.coef_, model.intercept_
```
